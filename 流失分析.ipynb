## 异常值处理


```python
data.describe(include='all').T
```

```python
a1 = ['lowestprice', 'lowestprice_pre']
def block_lower(x):
    ql = x.quantile(.01)
    out = x.mask(x < ql, ql)
    return(out)


def block_upper(x):
    qu = x.quantile(.99)
    out = x.mask(x > qu, qu)
    return(out)


data.loc[:, a1] = data.loc[:, a1].apply(block_upper)
data.loc[:, a1] = data.loc[:, a1].apply(block_lower)
data.loc[:, ['lowestprice', 'lowestprice_pre']].describe(include='all').T
```

### 利用随机森林选择特征

```python
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import cross_val_score
RFC_clf = RFC(n_estimators=10,random_state=0)
X_embedded = SelectFromModel(RFC_clf,threshold=0.05).fit_transform(Xtrain,Ytrain)
importance = RFC_clf.fit(Xtrain,Ytrain).feature_importances_
threshold = np.linspace(0,importance.max(),20)
score = []
for i in threshold:
    X_embedded = SelectFromModel(RFC_clf,threshold=i).fit_transform(Xtrain,Ytrain)
    val = cross_val_score(RFC_clf,X_embedded,Ytrain,cv=5).mean()
    score.append(val)
plt.plot(threshold,score)
plt.show()
```

```python
a=threshold[np.argmax(score)] #最高分对应的重要性阈值
```

```python
X_embedded = SelectFromModel(RFC_clf,threshold=a).fit_transform(Xtrain,Ytrain)
```

```python
col_k = list(Xtrain.columns[RFC_clf.feature_importances_>a])
col_k
```

```python
print('筛选出{}个特征'.format(len(col_k)))
```

```python
Xtrain_new = pd.DataFrame(X_embedded,columns = col_k )
Xtrain_new.info()
```

```python
Xtest_new = Xtest.loc[:,col_k].reset_index(drop=True)
Xtest_new
```

## 分箱

```python
f_bins = [-1,3,16,711]
m_bins = [8,246,544,6383]
r_bins = [-1,14926,22313.00,527026.0]
```

```python
rfm['R_score'] = pd.cut(rfm['R'],r_bins,labels=[i for i in range(len(r_bins)-1,0,-1)])
rfm['F_score'] = pd.cut(rfm['F'],f_bins,labels=[i+1 for i in range(len(r_bins)-1)])
rfm['M_score'] = pd.cut(rfm['M'],m_bins,labels=[i+1 for i in range(len(r_bins)-1)])
```

```python
rfm
```

## 计算RFM得分

```python
rfm = rfm.apply(np.int32)
rfm['rfm_score'] = rfm['F']*weights_rfm[0] + rfm['M']*weights_rfm[1] +rfm['R']*weights_rfm[2]
rfm['R_score'] = rfm['R_score'].astype(np.str_)
rfm['F_score'] = rfm['F_score'].astype(np.str_)
rfm['M_score'] = rfm['M_score'].astype(np.str_)
rfm['rfm_group'] = rfm['R_score'].str.cat(rfm['F_score']).str.cat(rfm['M_score'])
```
